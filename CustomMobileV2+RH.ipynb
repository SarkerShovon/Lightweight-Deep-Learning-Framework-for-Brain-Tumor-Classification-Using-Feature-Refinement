```python
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from math import ceil
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout, BatchNormalization, Conv2D, Activation, Add, Input, DepthwiseConv2D, ReLU, ZeroPadding2D, SeparableConv2D
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input as mobilenet_preprocess
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report, precision_score, recall_score, f1_score, roc_curve, auc
from sklearn.preprocessing import label_binarize
from tensorflow.keras.models import load_model
# ===============================
# Paths to dataset directories
# ===============================
base_dir = '/kaggle/input/dataset/Dataset_split'
train_dir = os.path.join(base_dir, 'train')
validation_dir = os.path.join(base_dir, 'validation')
test_dir = os.path.join(base_dir, 'test')

assert os.path.exists(train_dir), f"Train directory not found: {train_dir}"
assert os.path.exists(validation_dir), f"Validation directory not found: {validation_dir}"
assert os.path.exists(test_dir), f"Test directory not found: {test_dir}"

# ===============================
# Parameters
# ===============================
batch_size = 32
epochs = 150
num_classes = 4  # glioma, meningioma, notumor, pituitary

checkpoint_dir = '/kaggle/working/checkpoints'
plot_dir = '/kaggle/working/plots'
misclassified_dir = '/kaggle/working/misclassified_images'
os.makedirs(checkpoint_dir, exist_ok=True)
os.makedirs(plot_dir, exist_ok=True)
os.makedirs(misclassified_dir, exist_ok=True)

# ===============================
# Data Generators
# ===============================
def get_data_generators(preprocess_func, target_size):
    train_datagen = ImageDataGenerator(
        preprocessing_function=preprocess_func,
        rotation_range=20,
        width_shift_range=0.1,
        height_shift_range=0.1,
        shear_range=0.1,
        zoom_range=0.1,
        horizontal_flip=True,
        brightness_range=[0.8, 1.2],
        fill_mode='nearest'
    )

    validation_datagen = ImageDataGenerator(preprocessing_function=preprocess_func)

    train_generator = train_datagen.flow_from_directory(
        train_dir,
        target_size=target_size,
        batch_size=batch_size,
        class_mode='categorical'
    )

    validation_generator = validation_datagen.flow_from_directory(
        validation_dir,
        target_size=target_size,
        batch_size=batch_size,
        class_mode='categorical'
    )

    test_generator = validation_datagen.flow_from_directory(
        test_dir,
        target_size=target_size,
        batch_size=batch_size,
        class_mode='categorical',
        shuffle=False
    )

    return train_generator, validation_generator, test_generator

# ===============================
# Build Model
# ===============================
def inverted_residual_block(x, expansion, filters, stride, block_id, use_residual=True):
    prefix = f'block_{block_id}_'
    in_channels = x.shape[-1]
    input_tensor = x

    # Special case for the first block where expansion is not applied
    if block_id != 0:
        # Expansion phase
        x = Conv2D(expansion * in_channels, kernel_size=1, padding='same', use_bias=False, name=prefix + 'expand')(x)
        x = BatchNormalization(name=prefix + 'expand_BN')(x)
        x = ReLU(6., name=prefix + 'expand_relu')(x)

    # Depthwise Convolution
    if stride == 2:
        x = ZeroPadding2D(padding=((1, 1), (1, 1)), name=prefix + 'pad')(x)
    x = DepthwiseConv2D(kernel_size=3, strides=stride, padding='same' if stride == 1 else 'valid', use_bias=False, name=prefix + 'depthwise')(x)
    x = BatchNormalization(name=prefix + 'depthwise_BN')(x)
    x = ReLU(6., name=prefix + 'depthwise_relu')(x)

    # Projection phase
    x = Conv2D(filters, kernel_size=1, padding='same', use_bias=False, name=prefix + 'project')(x)
    x = BatchNormalization(name=prefix + 'project_BN')(x)

    # Residual connection
    if use_residual and in_channels == filters and stride == 1:
        x = Add(name=prefix + 'add')([input_tensor, x])

    return x

def MobileNetV2(input_shape=(224, 224, 3), num_classes=4):
    inputs = Input(shape=input_shape, name='input_layer')

    # Initial Conv layer
    x = Conv2D(32, kernel_size=3, strides=2, padding='same', use_bias=False, name='Conv1')(inputs)
    x = BatchNormalization(name='bn_Conv1')(x)
    x = ReLU(6., name='Conv1_relu')(x)

    # Define the architecture based on the summary
    x = inverted_residual_block(x, expansion=1, filters=16, stride=1, block_id=0, use_residual=False)
    x = inverted_residual_block(x, expansion=6, filters=32, stride=2, block_id=1, use_residual=False)
    x = inverted_residual_block(x, expansion=6, filters=32, stride=1, block_id=2)
    x = inverted_residual_block(x, expansion=6, filters=64, stride=2, block_id=3, use_residual=False)
    x = inverted_residual_block(x, expansion=6, filters=64, stride=1, block_id=4)
    x = inverted_residual_block(x, expansion=6, filters=64, stride=1, block_id=5)
    x = inverted_residual_block(x, expansion=6, filters=128, stride=2, block_id=6, use_residual=False)
    x = inverted_residual_block(x, expansion=6, filters=128, stride=1, block_id=7)
    x = inverted_residual_block(x, expansion=6, filters=128, stride=1, block_id=8)
    x = inverted_residual_block(x, expansion=6, filters=128, stride=1, block_id=9)
    x = inverted_residual_block(x, expansion=6, filters=256, stride=2, block_id=13, use_residual=False)
    

    # Final layers
    x = Conv2D(512, kernel_size=1, use_bias=False, name='Conv_1')(x)
    x = BatchNormalization(name='Conv_1_bn')(x)
    outputs = ReLU(6., name='out_relu')(x)

    

    model = Model(inputs, outputs, name='mobilenetv2')
    return model

# Load the custom MobileNetV2 model
base_model = MobileNetV2(input_shape=(224, 224, 3), num_classes=num_classes)
base_model.trainable = True

# Add custom classification head
x = base_model.output

# First convolutional block
x1 = Conv2D(64, (3, 3), padding='same')(x)
x1 = BatchNormalization()(x1)
x1 = Activation('relu')(x1)

# Second convolutional block
x2 = Conv2D(64, (3, 3), padding='same')(x1)
x2 = BatchNormalization()(x2)
x2 = Activation('relu')(x2)

# Residual connection
x = Add()([x1, x2])

# Additional convolutional layer
x1 = Conv2D(32, (3, 3), padding='same')(x)
x1 = BatchNormalization()(x1)
x1 = Activation('relu')(x1)


x2 = Conv2D(32, (3, 3), padding='same')(x1)
x2 = BatchNormalization()(x2)
x2 = Activation('relu')(x2)

# Residual connection
x = Add()([x1, x2])


x = SeparableConv2D(128, (3, 3), padding='same')(x)
x = BatchNormalization()(x)
x = Activation('relu')(x)

# Global average pooling
x = GlobalAveragePooling2D()(x)
# x = BatchNormalization()(x)

# Fully connected layer
x = Dense(512, activation='relu', kernel_regularizer='l2')(x)
x = Dropout(0.4)(x)  # Dropout for regularization

# Output layer
x = Dense(num_classes, activation='softmax')(x)

# Build the complete model
model = Model(inputs=base_model.input, outputs=x)
model.summary()

# ===============================
# Compile & Train
# ===============================
model.compile(optimizer=Adam(learning_rate=0.0001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

train_generator, validation_generator, test_generator = get_data_generators(
    mobilenet_preprocess, (224, 224)
)

checkpoint = ModelCheckpoint(
    filepath=os.path.join(checkpoint_dir, "MobileNetV2.keras"),
    save_best_only=True,
    monitor='val_loss'
)

reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.2,
    patience=5,
    min_lr=1e-6
)

history = model.fit(
    train_generator,
    epochs=epochs,
    validation_data=validation_generator,
    callbacks=[checkpoint, reduce_lr],
)

# ===============================
# Evaluation
# ===============================
best_model_path = os.path.join(checkpoint_dir, "MobileNetV2.keras")
best_model = load_model(best_model_path)

test_loss, test_acc = best_model.evaluate(test_generator)
print(f"Test accuracy for MobileNetV2: {test_acc:.4f}")

y_pred = best_model.predict(test_generator)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true = test_generator.classes

# Confusion matrix
conf_matrix = confusion_matrix(y_true, y_pred_classes)
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=test_generator.class_indices.keys())
disp.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.show()

# Classification Report (dict for per-class metrics)
class_labels = list(test_generator.class_indices.keys())
report = classification_report(y_true, y_pred_classes, target_names=class_labels, output_dict=True)
print(classification_report(y_true, y_pred_classes, target_names=class_labels))

# Per-class metrics including accuracy
print("\n=== Per-Class Metrics ===")
for label in class_labels:
    idx = class_labels.index(label)
    precision = report[label]['precision']
    recall = report[label]['recall']
    f1 = report[label]['f1-score']
    support = report[label]['support']

    class_indices_true = np.where(y_true == idx)[0]
    correct_predictions = np.sum(y_pred_classes[class_indices_true] == y_true[class_indices_true])
    class_accuracy = correct_predictions / len(class_indices_true)

    print(f"{label}: Precision={precision:.4f}, Recall={recall:.4f}, F1-Score={f1:.4f}, Accuracy={class_accuracy:.4f}")

# Weighted Precision, Recall, F1
precision = precision_score(y_true, y_pred_classes, average='weighted')
recall = recall_score(y_true, y_pred_classes, average='weighted')
f1 = f1_score(y_true, y_pred_classes, average='weighted')

print(f"\nWeighted Metrics -> Precision: {precision:.4f}, Recall: {recall:.4f}, F1-Score: {f1:.4f}")

# ===============================
# ROC Curves per Class
# ===============================
y_true_bin = label_binarize(y_true, classes=np.arange(num_classes))

plt.figure(figsize=(10, 8))
for i, label in enumerate(class_labels):
    fpr, tpr, _ = roc_curve(y_true_bin[:, i], y_pred[:, i])
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, lw=2, label=f'{label} (AUC = {roc_auc:.2f})')

plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve per Class')
plt.legend(loc="lower right")
plt.show()

# ===============================
# Plot Training History
# ===============================
def plot_history(history):
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    epochs_range = range(len(acc))

    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, acc, label='Training Accuracy')
    plt.plot(epochs_range, val_acc, label='Validation Accuracy')
    plt.legend(loc='lower right')
    plt.title('Training and Validation Accuracy')

    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, loss, label='Training Loss')
    plt.plot(epochs_range, val_loss, label='Validation Loss')
    plt.legend(loc='upper right')
    plt.title('Training and Validation Loss')

    plt.savefig(os.path.join(plot_dir, 'MobileNetV2_training_plot.png'))
    plt.show()

plot_history(history)

# ===============================
# Misclassified Images
# ===============================
filenames = test_generator.filenames
misclassified_indices = np.where(y_pred_classes != y_true)[0]
print(f"Total misclassified images: {len(misclassified_indices)}")

images_per_page = 9
num_pages = ceil(len(misclassified_indices) / images_per_page)

for page in range(num_pages):
    start = page * images_per_page
    end = start + images_per_page
    subset_indices = misclassified_indices[start:end]

    plt.figure(figsize=(15, 12))
    for idx, i in enumerate(subset_indices):
        img_path = os.path.join(test_dir, filenames[i])
        img = mpimg.imread(img_path)

        true_label = class_labels[y_true[i]]
        predicted_label = class_labels[y_pred_classes[i]]

        plt.subplot(3, 3, idx + 1)
        plt.imshow(img, cmap='gray')
        plt.title(f"True: {true_label}\nPred: {predicted_label}")
        plt.axis('off')

        save_path = os.path.join(misclassified_dir, f"{true_label}_as_{predicted_label}_{os.path.basename(filenames[i])}")
        plt.imsave(save_path, img, cmap='gray')

    plt.suptitle(f"Misclassified Images (Page {page + 1}/{num_pages})", fontsize=16)
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()
```
